```python
import pandas as pd
cars = pd.read_csv("imports-85.data")
cars.head()
cars.columns
```

```python
cars.columns = ["symboling","normalized_losses","make","fuel_type","aspiration","num_of_door","body_style","drive_wheels","engine_location","wheel_base","length","width","height","curb_weight","engine_type","num_of_cylinders","engine_size","fuel_system","bore","stroke","compression_ratio","horsepower","peak_rpm","city_mpg","highway_mpg","price"]
cars.columns
```

```python
cars.head()
```

```python
cars.info()
```

"price" column is the target column. We can use symboling,wheel_base,length,width,height,curb_weight,engine_size,compression_ratio,city_mpg,highway_mpg columns as features. 

```python
cars.iloc[0]
```

```python
import numpy as np
cars = cars.replace("?",np.nan)
cars.head()
```

```python
cars.dtypes
```

```python
cars[["normalized_losses","bore","stroke","horsepower","peak_rpm","price"]] = cars[["normalized_losses","bore","stroke","horsepower","peak_rpm","price"]].astype(float)
```

```python
cars.info()
```

```python
cars = cars.drop(["normalized_losses"],axis=1)
cars.info()
```

```python
cars = cars.dropna(axis=0)
cars.info()
```

```python
norm_cars = cars[["symboling","wheel_base","length","width","height","curb_weight","engine_size","bore","stroke","compression_ratio","horsepower","peak_rpm","city_mpg","highway_mpg","price"]]
norm_cars.head()
```

```python
norm_cars = np.absolute((norm_cars - norm_cars.mean()) / norm_cars.std())
norm_cars["price"] = cars["price"]
norm_cars
```

```python
norm_cars.columns
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(train_col,target_col,df):
    train = df.iloc[0:150]
    test = df.iloc[150:]
    knn = KNeighborsRegressor()
    knn.fit(train[[train_col]],train[target_col])
    predictions = knn.predict(test[[train_col]])
    mse = mean_squared_error(test[target_col],predictions)
    rmse = mse ** .5
    return rmse

price1 = knn_train_test("horsepower","price",norm_cars)
price2 = knn_train_test('engine_size',"price",norm_cars)
price3 = knn_train_test("bore","price",norm_cars)
price4 = knn_train_test('peak_rpm',"price",norm_cars)
price5 = knn_train_test('compression_ratio',"price",norm_cars)
price6 = knn_train_test('wheel_base',"price",norm_cars)
price7 = knn_train_test('symboling',"price",norm_cars)
price8 = knn_train_test('length',"price",norm_cars)
price9 = knn_train_test('width',"price",norm_cars)
price10 = knn_train_test('height',"price",norm_cars)
price11 = knn_train_test('curb_weight',"price",norm_cars)
price12 = knn_train_test('stroke',"price",norm_cars)
price13 = knn_train_test('city_mpg',"price",norm_cars)
price14 = knn_train_test('highway_mpg',"price",norm_cars)


print("price1",price1)
print("price2",price2)
print("price3",price3)
print("price4",price4)
print("price5",price5)
print("price6",price6)
print("price7",price7)
print("price8",price8)
print("price9",price9)
print("price10",price10)
print("price11",price11)
print("price12",price11)
print("price13",price11)
print("price14",price11)
```

```python
from sklearn.model_selection import cross_val_score, KFold

num_folds = [3, 5, 7, 9, 10, 11, 13, 15, 17, 19, 21, 23]

for fold in num_folds:
    kf = KFold(fold, shuffle=True, random_state=1)
    model = KNeighborsRegressor()
    mses = cross_val_score(model, norm_cars[['compression_ratio',"symboling","engine_size",'wheel_base','highway_mpg']], norm_cars["price"], scoring="neg_mean_squared_error", cv=kf)
    rmses = np.sqrt(np.absolute(mses))
    avg_rmse = np.mean(rmses)
    std_rmse = np.std(rmses)
    print(str(fold), "folds: ", "avg RMSE: ", str(avg_rmse), "std RMSE: ", str(std_rmse))
```
